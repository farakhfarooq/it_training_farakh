**First Task:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated stock prices
days = np.arange(1, 31)
prices = np.cumsum(np.random.randn(30) * 2) + 100  # Ensure correct formula

# Create a figure with specific size and resolution
plt.figure(figsize=(10, 5), dpi=100)

# Plot the stock price trend with markers
plt.plot(days, prices, linestyle='-', marker='o', color='b', markersize=6, label="Stock Price")

# Adding labels, title, and legend
plt.xlabel("Day")
plt.ylabel("Price (USD)")
plt.title("Stock Price Trend")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

**Second Task:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random temperature data
days = np.arange(1, 11)  # Days from 1 to 10
temp_day = np.random.randint(25, 35, size=10)  # Daytime temperatures (25°C to 35°C)
temp_night = np.random.randint(15, 25, size=10)  # Nighttime temperatures (15°C to 25°C)

# Create the figure with specific size
plt.figure(figsize=(8, 5))

# Plot Day Temperature with custom styles
plt.plot(days, temp_day, linestyle='--', marker='o', color='r', markersize=8, label="Day Temperature")

# Plot Night Temperature with custom styles
plt.plot(days, temp_night, linestyle='-', marker='s', color='b', markersize=8, label="Night Temperature")

# Adding labels, title, and legend
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Variation")

# Display legend
plt.legend()

# Display grid for better visualization
plt.grid(True)

# Show the plot
plt.show()
```

**Task no:03**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
countries = ["USA", "China", "India", "Brazil", "Germany"]
pop_growth = np.array([0.8, 1.2, 1.5, 0.9, 0.5])  # Growth in percentage
population = np.array([331, 1441, 1380, 213, 83])  # Population in millions

# Create a scatter plot with size variation
plt.figure(figsize=(8, 5))
scatter = plt.scatter(pop_growth, population, c=population, cmap='viridis',
                      s=population * 2, alpha=0.6, edgecolors="black")

# Labels
plt.xlabel("Population Growth (%)")
plt.ylabel("Total Population (millions)")
plt.title("Population Growth vs. Total Population")

# Adding annotations for each country
for i, country in enumerate(countries):
    plt.annotate(country, (pop_growth[i], population[i]), fontsize=10, ha='right')

# Add a colorbar to indicate population size
plt.colorbar(scatter, label="Population (millions)")

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

**Task no 04:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated customer data
np.random.seed(42)  # Ensures reproducibility
age = np.random.randint(18, 60, 100)  # Random ages between 18 and 60
spending_score = np.random.randint(20, 100, 100)  # Spending scores between 20 and 100
categories = np.random.choice([1, 2, 3, 4], 100)  # 4 different customer types

# Create a scatter plot with category-based colors
plt.figure(figsize=(8, 5))
scatter = plt.scatter(age, spending_score, c=categories, cmap='jet',
                      alpha=0.7, edgecolors="black")

# Adding labels and title
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation Analysis")

# Adding color bar for customer categories
plt.colorbar(scatter, label="Customer Category")

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

**Task no 05**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample sales data
np.random.seed(42)  # Ensures reproducibility
region = np.random.randint(1, 6, 50)  # 5 regions (1 to 5)
sales = np.random.randint(500, 10000, 50)  # Sales amount between 500 and 10,000
profit = np.random.randint(50, 1000, 50)  # Profit between 50 and 1,000

# Create a 3D figure
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(region, sales, profit, c=profit, cmap='coolwarm',
                     s=50, alpha=0.7, edgecolors="black")

# Adding labels
ax.set_xlabel("Region")
ax.set_ylabel("Sales")
ax.set_zlabel("Profit")
ax.set_title("3D Sales Performance Analysis")

# Adding a color bar for profit visualization
fig.colorbar(scatter, label="Profit")

# Show the plot
plt.show()
```
